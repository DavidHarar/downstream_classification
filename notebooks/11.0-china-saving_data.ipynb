```python
import os
import pandas as pd
import numpy as np
os.chdir('/home/david/Desktop/projects/thesis/')
data_folder_path = '/home/david/Desktop/projects/thesis/downstream_classification/data/china-raw'

files = list(os.listdir(data_folder_path))
files.sort()
files[:10]
```

```python
files  = list({x.split('.')[0] for x in files})
files[:10]
```

```python
import wfdb
import os
import numpy as np
import scipy.io

def load_single_sample(data_dir, record_name):
    # Construct the file names for the .hea and .mat files
    hea_file = os.path.join(data_dir, f"{record_name}.hea")
    mat_file = os.path.join(data_dir, f"{record_name}.mat")

    # Load the data from the .hea file
    try:
        record = wfdb.rdrecord(hea_file[:-4])  # Remove .hea for the record name
        data = record.p_signal  # Signal data
    except Exception as e:
        print(f"Error loading {hea_file}: {e}")
        return None, None

    # Load the corresponding label from the .mat file
    try:
        mat_data = scipy.io.loadmat(mat_file)
        # Adjust based on the actual structure of your .mat file
        labels = mat_data['label'][0]  # Assuming 'label' is the key in the .mat file
    except Exception as e:
        print(f"Error loading {mat_file}: {e}")
        return data, None

    return data, labels

# Usage example
for f in files:
    data, labels = load_single_sample(data_folder_path, f)

    if data is not None:
        print(f"Loaded data for {f} with shape: {data.shape}")
        print(f"Loaded labels: {labels}")
    else:
        print(f"Failed to load data for {f}.")
```
